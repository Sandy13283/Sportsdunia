```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Fill missing values with 'Unknown'
df['Registration Device'].fillna('Unknown', inplace=True)
```


# Plot the distribution of leads by State


```python
plt.figure(figsize=(10, 6))
sns.countplot(y='State', data=df, order=df['State'].value_counts().index)
plt.title('Distribution of Leads by State ')
plt.xlabel('Number of Leads')
plt.ylabel('State')
plt.show()
```

```python
df['State'].value_counts()
```

```python
df['City'].value_counts()
```

```python
plt.figure(figsize=(20, 6))
sns.countplot(y='City', data=df, order=df['City'].value_counts().head(10).index)
plt.title('Distribution of Leads by City ')
plt.xlabel('Number of Leads')
plt.ylabel('City')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Lead Stage', data=df, order=df['Lead Stage'].value_counts().index)
plt.title('Lead Stage Distribution')
plt.xlabel('Lead Stage')
plt.ylabel('Number of Leads')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Campaign', data=df, order=df['Campaign'].value_counts().head(10).index)
plt.title('Number of Leads by Campaign')
plt.xlabel('Campaign')
plt.ylabel('Number of Leads')
plt.xticks(rotation=45)
plt.show()
```

```python
lead_enrollment_ct = pd.crosstab(df['Lead Status'], df['Enrolment Status'])

# Plot the crosstab
lead_enrollment_ct.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Lead Status vs Enrolment Status')
plt.xlabel('Lead Status')
plt.ylabel('Number of Leads')
plt.show()
```

```python
df['Conversion Rate'] = (df['Paid Applications'] / df['Submitted Applications']).astype(float)

avg_conversion = df.groupby('Campaign')['Conversion Rate'].mean().sort_values(ascending=False).head(30)

plt.figure(figsize=(20, 12))
sns.barplot(x=avg_conversion.index, y=avg_conversion.values)
plt.title('Average Conversion Rate by Campaign (Sorted)')
plt.xlabel('Campaign')
plt.ylabel('Average Conversion Rate')
plt.xticks(rotation=90)
plt.show()
```

# Calculate the number of days between Instance Date and Last Lead Activity Date

```python
df['Days to Last Activity'] = (df['Last Lead Activity Date'] - df['Instance Date']).dt.days

plt.figure(figsize=(10, 6))
sns.histplot(df['Days to Last Activity'].sort_values(ascending=False), bins=30, kde=True)
plt.title('Distribution of Days to Last Activity')
plt.xlabel('Days to Last Activity')
plt.ylabel('Number of Leads')
plt.show()
```

```python
df.info()
```

```python
sns.barplot(x='Lead Stage', y='Days to Last Activity', data=df)
plt.title('Days to Last Activity by Lead Source')
plt.xlabel('Lead Source')
plt.ylabel('Days to Last Activity')
plt.xticks(rotation=45)
plt.show()
```
